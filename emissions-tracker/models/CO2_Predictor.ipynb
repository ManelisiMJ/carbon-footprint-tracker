```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("CarbonEmission.csv")

data = data.drop(['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11'], axis=1)
data= data.dropna()
data
```

```python
from sklearn.model_selection import train_test_split
x= data.drop(['TotalCarbonEmmission','EmmissionFromPurchasedElectricity','Company','Weeks'], axis=1)
y= data['TotalCarbonEmmission']
```

```python
x
```

```python
x_train ,x_test , y_train ,y_test = train_test_split(x, y , test_size=0.25)
```

```python
train_data = x_train.join(y_train)
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.corr()
```

```python
print(train_data.isnull().sum())  
```

```python
plt.figure()
train_data

# Sample a subset of the data
sample_size = 100  # Adjust the sample size according to your preference
sampled_data = train_data.sample(n=sample_size, random_state=42)

# Create the correlation matrix and heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(sampled_data.corr(), annot=True, cmap='YlGnBu')
```

```python
train_data['EmmissionFromIndirectSources']=np.log(train_data['EmmissionFromIndirectSources']+1)
train_data['CapitalGoods']=np.log(train_data['CapitalGoods']+1)
train_data['PowerConsumption(kWh)']=np.log(train_data['PowerConsumption(kWh)']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
from sklearn.linear_model import LinearRegression

x_train, y_train = train_data.drop(['TotalCarbonEmmission'], axis=1), train_data['TotalCarbonEmmission']
x_train
y_train
```

```python
reg = LinearRegression ()
reg.fit(x_train, y_train)
```

```python
train_data
```

```python
test_data = x_test.join(y_test)

test_data['EmmissionFromIndirectSources']=np.log(test_data['EmmissionFromIndirectSources']+1)
test_data['CapitalGoods']=np.log(test_data['CapitalGoods']+1)
test_data['PowerConsumption(kWh)']=np.log(test_data['PowerConsumption(kWh)']+1)
x_test, y_test = test_data.drop(['TotalCarbonEmmission'], axis=1), test_data['TotalCarbonEmmission']
```

```python
x_test
```

```python
x_train
```

```python
reg.score(x_test,y_test)
```
